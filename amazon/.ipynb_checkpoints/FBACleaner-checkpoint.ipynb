```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('rawdata.csv')
df
```

```python
#count the number of X's in dimensions
def countx(dim):
    xs = int(str(dim).count('x'))
    return xs

df['xs'] = df.apply(lambda row: countx(row['dimensions']), axis = 1)

```

```python
#filter by 2 x's

df1 = df[df['xs'] == 2]
df1

df1['l'] = df1['dimensions'].apply(lambda x: x.split(' ')[0])
df1['w'] = df1['dimensions'].apply(lambda x: x.split(' ')[2])
df1['h'] = df1['dimensions'].apply(lambda x: x.split(' ')[4])

df1
```

```python
def weightconverter(weight):
    num = float(str(weight).split(' ')[0])
    unit = str(weight).split(' ')[1]
    if unit == "ounces":
        lbs = num / 16
        return lbs
    else:
        return num

df1['lbs'] = df1.dropna().apply(lambda row: weightconverter(row['weight']), axis = 1)

df1
```

```python
df1['price'] = (df1['price'].replace( '[\$,)]','', regex=True )
               .replace( '[(]','-',   regex=True ).astype(float))


```

```python
df1['reviews']  = (df1['reviews'].dropna().replace( '[\$,)]','', regex=True )
                       .replace( '[(]','-',   regex=True ).astype(int))
df1['departmentRank']  = (df1['departmentRank'].dropna().replace( '[\$,)]','', regex=True ).astype(int))
```

```python
df1['l'] = (df1['l'].dropna().replace( '[\$,)]','', regex=True )
                       .replace( '[(]','-',   regex=True ).astype(float))
df1['w'] = (df1['w'].dropna().replace( '[\$,)]','', regex=True )
                       .replace( '[(]','-',   regex=True ).astype(float))
df1['h'] = (df1['h'].dropna().replace( '[\$,)]','', regex=True )
                       .replace( '[(]','-',   regex=True ).astype(float))

df1.dtypes
```

```python
df2 = df1[df1.price > 15] 
df3 = df2[df2.price < 40] 
df4 = df3[df3.lbs <= 1] 
df5 = df4[df4.reviews <= 500]
df6 = df5[df5.departmentRank > 500]
df7 = df6[df6.departmentRank < 10000]
df8 = df7[df7.l < 8]
df9 = df8[df7.w < 8]
df10 = df9[df7.h < 8]

df10

```

```python
df10.to_csv('filtered.csv')
```
